```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
dataset = pd.read_csv("UNSW_NB15_testing-set.csv")
dataset.head()

```

```python
# Preprocess the dataset
le = LabelEncoder()
dataset["attack_cat"] = le.fit_transform(dataset["attack_cat"])
dataset["proto"] = le.fit_transform(dataset["proto"])
dataset["service"] = le.fit_transform(dataset["service"])
scaler = StandardScaler()
dataset[["dur", "spkts", "dpkts", "sbytes", "dbytes", "rate", "sttl", "dttl", "sload", "dload", "sloss", "dloss", "sinpkt", "dinpkt", "sjit", "djit", "swin", "stcpb", "dtcpb", "dwin", "tcprtt", "synack", "ackdat"]] = scaler.fit_transform(dataset[["dur", "spkts", "dpkts", "sbytes", "dbytes", "rate", "sttl", "dttl", "sload", "dload", "sloss", "dloss", "sinpkt", "dinpkt", "sjit", "djit", "swin", "stcpb", "dtcpb", "dwin", "tcprtt", "synack", "ackdat"]])
```

```python
data = dataset.dropna()


nan_counts = data.isna().sum()
print(nan_counts)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Train a random forest classifier on the entire dataset
clf = RandomForestClassifier(n_estimators=100, random_state=42)
X = data.drop(columns=["attack_cat", "id", "label","state"])
y = data["attack_cat"]
clf.fit(X, y)

# Calculate the probability of each class for each sample
y_proba = clf.predict_proba(X)

# Calculate the trust score for each sample
trust_scores = np.zeros(len(X))
for i in range(len(X)):
    class_probs = y_proba[i]
    max_prob = np.max(class_probs)
    trust_scores[i] = max_prob / np.sum(class_probs)

# Add the trust scores to the dataset
data["trust_score"] = trust_scores

```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Define a threshold for the trust score
threshold = 0.9

@app.route("/", methods=["POST"])
def predict():
    # Get the input data from the request
    data = request.json

    # Preprocess the input data
    dur = (data["dur"] - dataset["dur"].mean()) / dataset["dur"].std()
    proto = le.transform([data["proto"]])[0]
    service = le.transform([data["service"]])[0]
    spkts = (data["spkts"] - dataset["spkts"].mean()) / dataset["spkts"].std()
    dpkts = (data["dpkts"] - dataset["dpkts"].mean()) / dataset["dpkts"].std()
    sbytes = (data["sbytes"] - dataset["sbytes"].mean()) / dataset["sbytes"].std()
    dbytes = (data["dbytes"] - dataset["dbytes"].mean()) / dataset["dbytes"].std()
    rate = (data["rate"] - dataset["rate"].mean()) / dataset["rate"].std()
    sttl = (data["sttl"] - dataset["sttl"].mean()) / dataset["sttl"].std()
    dttl = (data["dttl"] - dataset["dttl"].mean()) / dataset["dttl"].std()
    sload = (data["sload"] - dataset["sload"].mean()) / dataset["sload"].std()
    dload = (data["dload"] - dataset["dload"].mean()) / dataset["dload"].std()
    sloss = (data["sloss"] - dataset["sloss"].mean()) / dataset["sloss"].std()
    dloss = (data["dloss"] - dataset["dloss"].mean()) / dataset["dloss"].std()
    sinpkt = (data["sinpkt"] - dataset["sinpkt"].mean()) / dataset["sinpkt"].std()
    dinpkt = (data["dinpkt"] - dataset["dinpkt"].mean()) / dataset["dinpkt"].std()
    sjit = (data["sjit"] - dataset["sjit"].mean()) / dataset["sjit"].std()
    djit = (data["djit"] - dataset["djit"].mean()) / dataset["djit"].std()
    swin = (data["swin"] - dataset["swin"].mean()) / dataset["swin"].std()
    stcpb = (data["stcpb"] - dataset["stcpb"].mean()) / dataset["stcpb"].std()
    dtcpb = (data["dtcpb"] - dataset["dtcpb"].mean()) / dataset["dtcpb"].std()
    dwin = (data["dwin"] - dataset["dwin"].mean()) / dataset["dwin"].std()
    tcprtt = (data["tcprtt"] - dataset["tcprtt"].mean()) / dataset["tcprtt"].std()
    synack = (data["synack"] - dataset["synack"].mean()) / dataset["synack"].std()
    ackdat = (data["ackdat"] - dataset["ackdat"].mean()) / dataset["ackdat"].std()

   # Make a prediction
    X = np.array([[dur, proto, service, spkts, dpkts, sbytes, dbytes, rate, sttl, dttl, sload, dload, sloss, dloss, sinpkt, dinpkt, sjit, djit, swin, stcpb, dtcpb, dwin, tcprtt, synack, ackdat]])
    y_pred = clf.predict(X)
    pred_label = le.inverse_transform(y_pred)[0]
    # Calculate the trust score
    y_prob = clf.predict_proba(X)[0]
    if pred_label == "normal":
      trust_score = 1 - y_prob.max()
    else:
      trust_score = y_prob.max()

# Return the prediction and trust score
    if trust_score > threshold:
      result = {"prediction": pred_label, "trust_score": trust_score, "anomaly": False}
    else:
      result = {"prediction": pred_label, "trust_score": trust_score, "anomaly": True}
    return jsonify(result)
```
